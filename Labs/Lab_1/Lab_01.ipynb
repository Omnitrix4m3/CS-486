## Histogram

```python
plt.hist(df.Overall)
plt.title("Player Overalls")
plt.xlabel("Overall")
plt.ylabel("Number of Players")
plt.show()
```

This histogram demonstrates a normal distribution of the Overall values of players as there is no discernible left or right tail to the diagram. The most common Overall value lies within the 60-70 range. This suggests that there are very few poor or exceptionally gifted players, with the majority being less-than average to average.

## Correlation

```python
correlation = df.corr()
correlation
```

### Positive Correlations

```python
correlation.loc["Overall", "Wage"]
```

The positive correlation between a player's Overall value and their Wage value makes sense as you would traditionally expect the higher the skill of a player, the more they are compensated. As is true for the converse.

### Negative Correlations

```python
correlation.loc["GKDiving", "LongShots"]
```

When looking at the relationship between a player's GKDiving ability and their respective LongShots value, there is a negative correlation. This makes sense as someone that possesses higher diving ability, namely a goalkeeper, should have lower outfield player attributes such as shooting from distance, a skill which takes extensive practice that focuses more on attacking, rather than defending a goal.

```python
correlation.loc["Jersey Number", "SprintSpeed"]
```

The negative correlation between a player's Jersey Number and SprintSpeed is close enough to 0 to be considered statistically significant. This is because the number a player wears (an identity based characteristic) should have no bearing on their physical attributes.
